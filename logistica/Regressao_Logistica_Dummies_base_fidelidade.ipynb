## Dummizando

```python
fidelidade_dummies = pd.get_dummies(df_fidelidade, columns=['atendimento', 
                                                            'sortimento', 
                                                            'acessibilidade', 
                                                            'preço', 
                                                            'sexo'], 
                                    drop_first='True')
fidelidade_dummies
```

```python
fidelidade_dummies.loc[fidelidade_dummies['fidelidade'] == 'sim', 'target'] = int(1)
fidelidade_dummies.loc[fidelidade_dummies['fidelidade'] == 'nao', 'target'] = int(0)
```

```python
fidelidade_dummies
```

## Aplicando Modelo Logístico

O modelo que apresentar o MAIOR Log-Likelihood é o mais ajustado.

```python
import statsmodels.api as sm

# Caso for usar a função sem a formula
X = fidelidade_dummies.iloc[:,2:-1]
y = fidelidade_dummies.target

#Adiciona coluna para cálculo da constante
X = sm.add_constant(X)

modelo_fidelidade = sm.Logit(y,X).fit()

#Função com Formula
#modelo_challenger = sm.Logit.from_formula(formula='falha ~ temperatura + pressão', data=df_challenger).fit()
modelo_fidelidade.summary()
```

## Stepwise

preço_2 não passou ao nível de 95% de confiança com p-value de 0,788

```python
from stepwise_process.statsmodels import stepwise

step_fidelidade = stepwise(modelo_fidelidade, pvalue_limit=0.05)
step_fidelidade.summary()
```

## Indicadores

Compara modelos levando em consideração a quantidade do tamanho da amostra e variáveis

- #### Akaike Info Criterion: Quanto menor Melhor
    - Análogo ao R2 Ajustado
    - AIC = -2 LLmodelo + 2(K+1) onde k = qtde de betas
    
- #### Bayesian Info Criterion: Quanto menor Melhor
    - BIC = -2LLmodelo + (k+1)ln(n)

```python
list_ = [step_fidelidade.llr.round(3), 
         step_fidelidade.llr_pvalue.round(5),
         step_fidelidade.llf.round(3),
         step_fidelidade.aic,
         step_fidelidade.bic]

summary = pd.DataFrame(list_, index=[['chi2', 'p', 'Log-Likelihood','AIC', 'BIC']], columns=['values'])

summary
```

## Confusion Matrix e Cutoff

Com Cutoff de 50%

Taxa média de cutoff no Brasil no crédito de financiamento: 22%


Períodos com taxa de juros subindo o cutoff pode subir associado à taxa de juros

```python
from sklearn.metrics import confusion_matrix

cutoff = 0.5

predicted_classes = (step_fidelidade.fittedvalues > cutoff).astype(int)

cm = pd.DataFrame(confusion_matrix(y, predicted_classes), 
                  index=['False', 'True'],
                  columns=['False', 'True'])
cm['Real'] = cm.T.sum()
```

## Curva ROC ( Receiver Operating Characterist)

Maior a área debaixo da curva ROC melhor. <br>

Para aumentar a área é necessário incluir novas variáveis com betas estatisticamente significantes após stepwise

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calcular a curva ROC e a área sob a curva (AUC)
fpr, tpr, thresholds = roc_curve(y, step_fidelidade.fittedvalues)
roc_auc = roc_auc_score(y, step_fidelidade.fittedvalues)

# Plotar a curva ROC
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Linha de referência
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falso Positivo')
plt.ylabel('Taxa de Verdadeiro Positivo')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
```
