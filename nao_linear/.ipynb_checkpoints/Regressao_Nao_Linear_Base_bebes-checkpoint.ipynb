## Contexto

 - Base contendo o comprimento em cm e a idade em semanas de vida;

```python
#Importa base e transforma em objeto
rData = pyreadr.read_r(r'bases\bebes.RData')

#Verifica o nomne do objeto
print(rData.keys())
```

```python
#Transforma o objeto em um pandas DataFrame
bebes = rData['bebes']
bebes.head()
```

```python
bebes.describe()
```

```python
#Cria a reta de regressão das dispersões
from scipy.optimize import curve_fit
from scipy import stats


# Cria a curva não linear
def func(x, a, b):
    return a * np.power(x, b)

popt, pcov = curve_fit(func, bebes.idade, bebes.comprimento)

x_fit = np.linspace(2, 60, 80)
y_fit = func(x_fit, * popt)

# Calcular a equação da reta linear
slope, intercept, r_value, p_value, std_err = stats.linregress(x=bebes.idade, y=bebes.comprimento)
y_linear = intercept+slope*bebes.idade

import plotly.graph_objs as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=bebes.idade, y=bebes.comprimento, mode='markers', name='Bebes'))
fig.add_trace(go.Scatter(x=x_fit, y=y_fit, mode='lines', name='Reta de regressão não linear'))
fig.add_trace(go.Scatter(x=bebes.idade, y=y_linear, mode='lines', name='Reta linear'))

fig.update_layout(
    title='Gráfico Comprimento x Idade dos bebês',
    xaxis_title='Idade',
    yaxis_title='Comprimento',
)

fig.show()
```

# Estimação do Modelo OLS Linear


Apesar do modelo ter apresentado um p-valor dentro do nível de significância, devemos verificar a aderência dos resíduos à normalidade.


```python
#Estimando o modelo
#https://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.OLS.html#statsmodels.regression.linear_model.OLS
import statsmodels.api as sm

modelo_linear = sm.OLS.from_formula(formula = 'comprimento ~ idade', data=bebes).fit()
modelo_linear.summary()
```

### Teste de Verificação da Aderência dos Resíduos à Normalidade (Shapiro-Francia)

Caso os erros não sigam uma distribuição normal, o modelo não é adequado para fins preditivos. <br>
Os betas não são ajustados para que consigam fazer inferência a partir dos intervalos de confiança daqueles parâmetros e construção de cenários.

> Teste Shapiro-Francia: n >= 30

> Teste Shapiro-Wilk: n < 30


onde,

 - H0: Diferênças não são estatisticamente significantes (existe aderência à normalidade);
 - H1: Diferênças são estatisticamente significantes (não é uma normal)
 
p-value < 0,05 rejeita-se H0

```python
from scipy.stats import shapiro

stat, p_value = shapiro(modelo_linear.resid)
stat , p_value
```

```python
modelo_linear.resid.hist()
```

### Transformação de Box-Cox

O modelo não possui resíduos aderentes à normalidade. Devemos transformar a variável Y em uma normal aplicando o método de Box Cox:


<center><img src='img\box_cox.jpg'></center>


- Lâmbda 1: linear
- Lâmbda 2: quadrática
- Lâmbda 3: cúbica
- Lâmbda -1: inversa
- Lâmbda 0,5: raiz
- Lâmbda -0,5: raiz inversa
- Lâmbda 0: ln(y

Lâmbda varia de -infinito a +infinito.

<b>Objetivo é encontrar qual o lâmbda que maximiza a aderência do Y à normalidade.</b>


<b> A padronização pelo Z-Score NÃO normaliza, a distribuição se mantém a mesma.<b>

```python
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox.html
from scipy import stats

y_lamb, lambida_otimo = stats.boxcox(bebes.comprimento)
lambida_otimo
```

```python
bebes['bc_comprimento'] = y_lamb
bebes
```

### Estimando Novo Modelo OLS com a Variável Dependente Transformada

<b> Os PARÂMETROS do modelo_linear e do modelo_bc NÂO são comparáveis!<br>
    Devemos fazer a transformação inversa usando o LAMBDA ÓTIMO </b>

```python
modelo_bc = sm.OLS.from_formula(formula = 'bc_comprimento ~ idade', data=bebes).fit()
modelo_bc.summary()
```

#### Existe um salto na qualidade do ajuste para o modelo não linear

```python
print(f'R2 Modelo Linear: {round(modelo_linear.rsquared, 3)}')
print(f'R2 Modelo Não Linear: {round(modelo_bc.rsquared, 3)}')
```

### Teste Shapiro-Francia para o Modelo Não Linear

```python
stat, p_value = shapiro(modelo_bc.resid)
stat , p_value
```

```python
modelo_bc.resid.hist()
```

### Comparando as Previsões dos Dois Modelo

```python
#https://www.statsmodels.org/dev/examples/notebooks/generated/ols.html

# Ajustando intercepto
x_train = sm.add_constant(bebes.idade)


modelo_linear = sm.OLS(endog=bebes.comprimento, exog=x_train).fit()
```

```python
# Ajustando intercepto
x_test = np.array([[1,19]])
x_test = sm.add_constant(x_test)
```

```python
modelo_linear.predict(x_test)
```

```python
modelo_bc = sm.OLS(endog=bebes.bc_comprimento, exog=x_train).fit()
```

```python
modelo_bc.predict(x_test)
```

#### Transformação Inversa

(Y(lamb) * lamb) +1 ^ (1/lamb)

### Inserindo Resultados na Base

```python
bebes['yhat_linear'] = modelo_linear.fittedvalues
bebes['yhat_bc'] = (((modelo_bc.fittedvalues * lambida_otimo) + 1)) ** (1/lambida_otimo)
bebes
```

```python
def func(x, a, b):
    return a * np.power(x, b)

popt_bc, pcov = curve_fit(func, bebes.comprimento, bebes.yhat_bc)

x_fit_bc = np.linspace(32, 88, 50)
y_fit_bc = func(x_fit_bc, *popt_bc)


popt_lin, pcov = curve_fit(func, bebes.comprimento, bebes.yhat_linear)

x_fit_lin = np.linspace(32, 88, 50)
y_fit_lin = func(x_fit_bc, *popt_lin)


# Calcular a equação da reta linear
slope, intercept, r_value, p_value, std_err = stats.linregress(x=bebes.comprimento, y=bebes.comprimento)
element_line = intercept+slope*bebes.comprimento


fig = go.Figure()
fig.add_trace(go.Scatter(x=bebes.comprimento, y=bebes.yhat_linear, mode='markers', name='yhat_linear'))
fig.add_trace(go.Scatter(x=bebes.comprimento, y=bebes.yhat_bc, mode='markers', name='yhat_bc'))
fig.add_trace(go.Scatter(x=x_fit_bc, y=y_fit_bc, mode='lines', name='Box-Cox'))
fig.add_trace(go.Scatter(x=x_fit_lin, y=y_fit_lin, mode='lines', name='OLS Linear'))
fig.add_trace(go.Scatter(x=bebes.comprimento, y=element_line, mode='lines', name='Ideal'))

fig.update_layout(
    title='Gráfico Comprimento x Idade dos bebês',
    xaxis_title='Comprimento',
    yaxis_title='Fitted Values',
)

fig.show()
```
